```python
def remove_plus(record):
    if(str(record).find("+") != -1):
        t=record.split("+")
        return int(t[0])+int(t[1])
```

```python
df_fifa[["gk_diving","gk_handling","gk_kicking","gk_reflexes","gk_speed","gk_positioning"]]=df_fifa[["gk_diving","gk_handling","gk_kicking","gk_reflexes","gk_speed","gk_positioning"]].fillna(0)
```

```python
def zero_check(x):
    if(x==0):
        return df_fifa['value_eur'].median()
    else:
        return x
```

```python
df_fifa['value_eur'] = df_fifa['value_eur'].apply(lambda x : zero_check(x))
```

```python
not_col = ['sofifa_id','short_name','club']
```

```python
df_col = df_fifa.columns.difference(not_col)
```

```python
df_final = df_fifa[df_col]
```

```python
def age_convert(x):
    x = int(x)
    if(x>15 and x<20):
        return 1
    elif(x>=20 and x<30):
        return 2
    elif(x>=30 and x<40):
        return 3
    elif(x>=40 and x<50):
        return 4
```

```python
df_final['age'].value_counts()
```

```python
df_final['age'] = [age_convert(x) for x in df_final['age']]
```

```python
categorical_col = ['nationality','player_positions','preferred_foot','age','body_type','international_reputation','skill_moves','weak_foot','work_rate']
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
```

```python
train = df_final[df_final.columns.difference(['value_eur','year'])]
test= df_final['value_eur']
```

```python
for col in categorical_col:
    train[col] = train[col].astype('category')
```

```python
train_x,test_x,train_y,test_y = train_test_split(train,test,test_size=0.2,random_state=2020)
```

```python
params = {
        "objective" : "regression",
        "metric" : "rmse",
        "num_leaves" : 40,
        "learning_rate" : 0.004,
        "bagging_fraction" : 0.6,
        "feature_fraction" : 0.6,
        "bagging_frequency" : 6,
        "bagging_seed" : 42,
        "verbose" : -1,
        "seed":42
    }
```

```python
lgb_train = lgb.Dataset(train_x,label=train_y)
lgb_val = lgb.Dataset(test_x,label=test_y)
evals_result = {}
model = lgb.train(params,lgb_train,10000,valid_sets=[lgb_train,lgb_val],
                 early_stopping_rounds=50,verbose_eval=150,
                 evals_result=evals_result,categorical_feature=categorical_col)
```

```python
gain = model.feature_importance('gain')
featureimp = pd.DataFrame({'feature':model.feature_name(),
                         'split':model.feature_importance('split'),
                         'gain': 100*gain/gain.sum()}).sort_values('gain',ascending=False)
```
